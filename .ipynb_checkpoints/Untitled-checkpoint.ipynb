### Loading Necessary Libraries

```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

### Basic Graph

```python
x=[0,1,2,3,4]
y=[0,2,4,6,8]
##for resizing
plt.figure(figsize=(5,3),dpi=130)

##use short hand notation
#fmt ='[color][marker][line]'
plt.plot(x,y,'b.--',label='2x')

##Line Number two
x2=np.arange(0,4.5,0.5)

#plot part of a graph as the line
plt.plot(x2[:5],x2[:5]**2,'r',label='x^2')

#plot remainder as the graph as a dot
plt.plot(x2[4:],x2[4:]**2,'r.-')

#Add A title(specify fontdict and its formats)
plt.title('Our First Graph!',fontdict={'fontname':'Brush Script MT','fontsize':18})

#X and Y labels
plt.xlabel("X axis",fontdict={'fontname':'Lucida Bright','fontsize':10})
plt.ylabel("Y axis",fontdict={'fontname':'Lucida Bright','fontsize':10})

#X ,Y tickmarks (scale of your graph)
plt.xticks([0,1,2,3,4])
# plt.yticks([0,2,4,6,8,10])

#Add a legend
plt.legend()

#Save Graph(dpi 300 is good when saving graphs as high resolution)
plt.savefig('matplotlibsample1.png',dpi=300)

#Show Plot
plt.show()
```

### Bar Charts

```python
labels=['A','B','C']
values=[1,4,2]
bars=plt.bar(labels,values)
patterns=['o','/','*']
for bar in bars:
    bar.set_hatch(patterns.pop(0))
plt.figure(figsize=(6,4),dpi=100)
plt.show()
```
